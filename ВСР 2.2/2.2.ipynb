# Проект: Числа Фибоначчи и золотое сечение

В этом проекте я:
- Сгенерировал первые n чисел Фибоначчи.
- Построил графики роста последовательности.
- Исследовал отношение соседних чисел и его приближение к золотому сечению (≈1.618).


```julia
using Plots
gr()   # явно говорим: использовать бэкенд GR
```

```julia
plot(1:10, rand(10), lw=2, marker=:circle, label="Тест")
```

```julia
# Функция для первых n чисел Фибоначчи
function fibonacci(n)
    fib = [0, 1]
    for i in 3:n
        push!(fib, fib[end] + fib[end-1])
    end
    return fib[1:n]
end

# Пример: первые 15 чисел
fibonacci(15)
```

```julia
fib20 = fibonacci(20)

# Строим график роста
plot(fib20, 
    lw=2, marker=:circle, label="Fibonacci", 
    title="Рост последовательности Фибоначчи",
    xlabel="Номер элемента", ylabel="Значение")
```

```julia
# Отношения Fn+1/Fn
ratios = [fib20[i+1]/fib20[i] for i in 2:length(fib20)-1]

plot(ratios,
    lw=2, marker=:star, color=:red,
    title="Отношение соседних чисел Фибоначчи",
    xlabel="Номер элемента", ylabel="Fn+1 / Fn",
    label="Отношение Fn+1/Fn")
hline!([1.618], label="Золотое сечение", linestyle=:dash)
```

## Выводы

- Последовательность Фибоначчи быстро растёт.
- Отношение соседних чисел стремится к золотому сечению (≈ 1.618).
- Даже простая последовательность демонстрирует красивое математическое свойство.
